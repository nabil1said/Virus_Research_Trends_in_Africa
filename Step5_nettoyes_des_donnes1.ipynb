```python
import time
from Bio import Entrez
import xml.etree.ElementTree as ET


Entrez.email = "toybngazi92@gmail.com"

pmid_list = [str(pmid) for pmid in pmid_list]

def get_affiliations_dict(pmids, batch_size=2, delay=5):
    result = {pmid: [] for pmid in pmids}

    for i in range(0, len(pmids), batch_size):
        batch = pmids[i:i+batch_size]
        try:
            print(f"Traitement du batch : {batch}")
            handle = Entrez.efetch(db="pubmed", id=",".join(batch), retmode="xml")
            root = ET.parse(handle).getroot()
            handle.close()

            articles = root.findall(".//PubmedArticle")
            for article in articles:
                pmid_element = article.find(".//PMID")
                if pmid_element is None:
                    continue
                pmid = pmid_element.text
                affiliations = set()

                for aff in article.findall(".//AffiliationInfo/Affiliation"):
                    if aff.text:
                        affiliations.add(aff.text.strip())

                if pmid in result:
                    result[pmid] = list(affiliations)

        except Exception as e:
            print(f"Erreur avec le batch {batch} : {e}")

        time.sleep(delay)

    return result

affiliations_dict = get_affiliations_dict(pmid_list)
```

```python
import pycountry
import unicodedata
from babel import Locale
from country_capitals import get_capital
from flashtext import KeywordProcessor
import pandas as pd

def normalize_string(text):
    """Normalise une chaîne de caractères en supprimant les accents et caractères spéciaux."""
    if not text:
        return ""
    return unicodedata.normalize('NFKD', text).encode('ASCII', 'ignore').decode('ASCII')

def create_country_dictionary():
    """Crée un dictionnaire contenant des informations sur les pays."""
    country_dict = {}
    locale_fr = Locale('fr')
    locale_en = Locale('en')

    for country in pycountry.countries:
        try:
            name_en = country.name
            official_name = getattr(country, 'official_name', name_en)
            iso3 = country.alpha_3.lower()
            capital = get_capital(name_en) or ""

            name_fr = locale_fr.territories.get(country.alpha_2, "")
            name_en_localized = locale_en.territories.get(country.alpha_2, "")

            normalized_name_fr = normalize_string(name_fr).upper()

            country_dict[normalized_name_fr] = [
                name_en.lower(),          # Nom en anglais (original pycountry)
                iso3,
                capital.lower(),
                name_fr.lower(),
                name_en_localized.lower()
            ]
        except Exception:
            continue

    return country_dict

# Création du dictionnaire complet
country_dictionary = create_country_dictionary()
```

```python
def create_country_dict_and_keyword_processor():
    # Créer un KeywordProcessor et ajouter les mots-clés
    keyword_processor = KeywordProcessor()
    for key in country_dictionary:
        keyword_processor.add_keyword(key.lower(), country_dictionary[key][0].lower())

    return country_dictionary, keyword_processor

# Utilisation de la fonction
country_dictionary, kp = create_country_dict_and_keyword_processor()

def detect_countries_as_string(text, keyword_processor):
    """Renvoie une chaîne des pays détectés, séparés par virgule"""
    # Convertir le texte en minuscules
    text_lower = str(text).lower()
    matches = keyword_processor.extract_keywords(text_lower)
    return ", ".join(sorted(set(matches))) if matches else ""

# Exemple d'application

country_dictionary["REPUBLIQUE CENTRAFRICAINE"] = [
    'Central African Republic',
    'Central African Republic',
    'CAF',
    'Bangui',
    'République centrafricaine', "centrafricaine", "Centrafrique"
]

# Enlever les espaces en début et fin + réduire les espaces multiples entre les mots à un seul espace
data_affiliation_na["Affiliation"] = data_affiliation_na["Affiliation"].str.strip().str.replace(r'\s+', ' ', regex=True)

data_affiliation_na["new_country"] = [detect_countries_as_string(x, kp) for x in data_affiliation_na["Affiliation"]]
```
