```python
import re

# Créer les colonnes si elles n'existent pas
df["Virus_Presence"] = ""
df["presence_dae"] = ""

for index, row in df.iterrows():
    # Vérification sécurisée : si ce n'est pas une liste, on considère vide
    viruses = row["Detected_Viruses"] if isinstance(row["Detected_Viruses"], list) else []
    genus = row["Detected_Genus"] if isinstance(row["Detected_Genus"], list) else []
    families = row["Detected_Families"] if isinstance(row["Detected_Families"], list) else []
    text = str(row["Title_Abstract"])  # pour éviter les erreurs si NaN

    # Cas "virus"
    if len(viruses) == 0 and len(genus) == 0 and re.search(r'\b[\w\-]*virus\b', text, re.IGNORECASE):
        df.at[index, "Virus_Presence"] = ", ".join(
            re.findall(r'\b[\w\-]*virus\b', text, re.IGNORECASE)
        )

    # Cas "dae"
    if len(viruses) == 0 and len(families) == 0 and re.search(r'\b[\w\-]*dae\b', text, re.IGNORECASE):
        df.at[index, "presence_dae"] = ", ".join(
            re.findall(r'\b[\w\-]*dae\b', text, re.IGNORECASE)
        )
```

# Recherche des virus  dans "Virus_Presence"

```python
# 1. Filtrer les lignes où 'Virus_Presence' n’est pas vide ou NaN
data_virus_presence = df[
    df['Virus_Presence'].notna() &
    (df['Virus_Presence'].str.strip() != "")
]

# 2. Nettoyer la colonne 'Virus_Presence'
data_virus_presence.loc[:, 'Virus_Presence'] = (
    data_virus_presence['Virus_Presence']
    .str.lower()  # Tout en minuscules
    .str.replace(r'\bvirus\b', '', regex=True)  # Enlève le mot 'virus' isolé
    .str.replace(r'\s*,\s*', ',', regex=True)   # Supprime les espaces autour des virgules
    .str.replace(r',+', ',', regex=True)        # Remplace les virgules multiples par une seule
    .str.strip(', ')                            # Supprime les virgules et espaces inutiles en début/fin
)

# 3. Refiltres les lignes si le nettoyage a vidé certaines cellules
data_virus_presence = data_virus_presence[
    data_virus_presence['Virus_Presence'].notna() &
    (data_virus_presence['Virus_Presence'].str.strip() != "")
]
```

```python
## normalisations  en transformant  en list et spurresion de doublons
data_virus_presence['Virus_Presence'] = data_virus_presence['Virus_Presence'].apply(
    lambda x: list(dict.fromkeys([item.strip() for item in x.split(',')]))
             if isinstance(x, str)
             else []
)
```

```python
data_virus_presence['Virus_Presence']
```

```python
dict_virus_variant = (df_taxons.groupby('Virus')['Variant']
                      .apply(lambda x: x.str.lower().unique().tolist())
                      .to_dict())
def netoyage_dic(nom_du_dictionnaire):
    for key, value in nom_du_dictionnaire.items():
        value = [mot.split(',') for mot in value]
        value = [mot.strip() for sous_liste in value for mot in sous_liste]
        nom_du_dictionnaire[key] = value
    return nom_du_dictionnaire


dict_virus_variant = netoyage_dic(dict_virus_variant)
```

```python
dict_virus_variant ["Zika virus"]
```

```python
# Étape 2 — Fonction de recherche dans le dictionnaire
def trouver_nom_standard(valeur, dico):
    if not isinstance(valeur, str):
        return None
    valeur = valeur.lower().strip()
    for nom_standard, variantes in dico.items():
        variantes_lower = [v.lower().strip() for v in variantes]
        if valeur in variantes_lower:
            return nom_standard
    return None  # Aucun nom standard trouvé

# Étape 3 — Mapper chaque liste à une liste de noms standard
def standardiser_virus(liste_virus):
    if not isinstance(liste_virus, list):
        return []

    virus_standards = []
    for v in liste_virus:
        nom = trouver_nom_standard(v, dict_virus_variant)
        if nom:  # ignore None
            virus_standards.append(nom)

    return list(set(virus_standards))  # Supprime les doublons

# Étape 4 — Appliquer au DataFrame pour créer une nouvelle colonne
data_virus_presence['Virus_standard'] = data_virus_presence['Virus_Presence'].apply(standardiser_virus)
```

```python


def merge_lists_simple(source_dataframe, target_dataframe, lists_column, result_column, key_column='PMID'):
    """
    Fusionne des listes d'un DataFrame source dans un DataFrame cible.

    Parameters:
    - source_dataframe (pd.DataFrame): Le DataFrame source contenant les listes à fusionner.
    - target_dataframe (pd.DataFrame): Le DataFrame cible dans lequel les listes seront fusionnées.
    - lists_column (str): Le nom de la colonne contenant les listes dans le DataFrame source.
    - result_column (str): Le nom de la colonne dans le DataFrame cible où les listes fusionnées seront stockées.
    - key_column (str): Le nom de la colonne utilisée comme clé pour la fusion. Par défaut, 'PMID'.
    """
    # Filtrer les lignes où la liste n'est pas vide
    source_dataframe = source_dataframe[source_dataframe[lists_column].apply(lambda x: len(x) > 0)]

    # Convertir la clé en string dans les deux DataFrames
    source_dataframe[key_column] = source_dataframe[key_column].astype(str)
    target_dataframe[key_column] = target_dataframe[key_column].astype(str)

    # Grouper et fusionner les listes
    grouped_dataframe = source_dataframe.groupby(key_column)[lists_column].agg(lambda lists: sum(lists, [])).reset_index()

    # Fusionner avec le DataFrame cible
    merged_dataframe = target_dataframe.merge(grouped_dataframe, on=key_column, how='left')

    # Remplacer NaN par liste vide
    merged_dataframe[lists_column] = merged_dataframe[lists_column].apply(lambda x: x if isinstance(x, list) else [])

    # Concaténer les listes dans la colonne résultat (en gérant si elle existe déjà ou pas)
    def concatenate_lists(row):
        base = row[result_column] if isinstance(row[result_column], list) else [row[result_column]] if pd.notna(row[result_column]) else []
        return list(dict.fromkeys(base + row[lists_column]))  # Enlever les doublons en conservant l'ordre

    merged_dataframe[result_column] = merged_dataframe.apply(concatenate_lists, axis=1)

    # Supprimer la colonne temporaire
    return merged_dataframe.drop(columns=[lists_column])
```
